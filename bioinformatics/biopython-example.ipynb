```python
from Bio import pairwise2

p53_human = "ACGGATGCCCGGTTTAAACCGGTTTTACGTACCGAATGCGT"
p53_mouse = "ACGGTTAAGGTTTAAACCGGTTTCCGTACCGAATGCGGCTT"


alns = pairwise2.align.globalms(p53_human, p53_mouse, 2, 0, -1, -0.5)

# print each of the alignments

for i in alns:
	print(i)
	
print("-----------------")

# work with the top scoring alignment

top_aln = alns[0]

# assigns each value in list to variables which can then individually be called

aln_human, aln_mouse, score, begin, end = top_aln

print (aln_human)
print (aln_mouse)
print (score)
print (end)

print (aln_human+'\n'+aln_mouse)

```

```python
#blast-nuc.py code

from Bio import SeqIO
from Bio.Blast import NCBIWWW

my_query = SeqIO.read("cftr1.txt", format="fasta")
result_handle = NCBIWWW.qblast("blastn", "nt", my_query.seq)
blast_result = open("my_blast.xml", "w")
blast_result.write(result_handle.read())
blast_result.close()
result_handle.close()

from Bio.Blast import NCBIXML

E_VALUE_THRESH = 1e-20

for record in NCBIXML.parse(open("my_blast.xml")):
    if record.alignments: #skip queries with no matches
        for align in record.alignments:
            for hsp in align.hsps:
                if hsp.expect < E_VALUE_THRESH:
                    print ("MATCH: %s " % align.title[:60])
                    print (hsp.expect)
                    print (hsp.query[:60])
                    print (hsp.match[:60])
                    print (hsp.sbjct[:60])
```

```python
#blasti-prot.py code

from Bio import SeqIO
from Bio.Blast import NCBIWWW

my_query = SeqIO.read("EF1a.fasta", format="fasta")
result_handle = NCBIWWW.qblast("blastp", "pdb", my_query.seq)
blast_result = open("my_blast.xml", "w")
blast_result.write(result_handle.read())
blast_result.close()
result_handle.close()

from Bio.Blast import NCBIXML

E_VALUE_THRESH = 1e-20

for record in NCBIXML.parse(open("my_blast.xml")):
	if record.alignments: #skip queries with no matches
		for align in record.alignments:
			for hsp in align.hsps:
				if hsp.expect < E_VALUE_THRESH:
					print ("MATCH: %s " % align.title[:60])
					print ("LENGTH:", hsp.align_length)
					print ("IDENTITIES:", hsp.identities)
					print ("SCORE:", hsp.score)
					print ("GAPS:", hsp.gaps)
					print (hsp.query[0:80])
					print (hsp.match[0:80])
					print (hsp.sbjct[0:80])
```
