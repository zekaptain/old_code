```python
import pandas
titanicData = pandas.read_csv('titanicFull.csv')

meanAge = titanicData['Age'].mean() #super easily prints the mean age in one fucking line

titanicData['Age'] = titanicData['Age'].fillna( meanAge )

zScore = (titanicData['Age'] - titanicData['Age'].mean()) / titanicData['Age'].std()
#print zScore

titanicData['Age'] = zScore #this replaces every val in Age with its normalized z-Score
#titanicData
```
