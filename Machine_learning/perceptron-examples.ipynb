```python
import pandas
from sklearn import linear_model as lm
from sklearn import cross_validation as cv
from sklearn import metrics
from sklearn import preprocessing as pp

iris_data = data = pandas.read_csv('irisData.csv')

(iris_train,iris_test) = cv.train_test_split(iris_data,test_size=0.2)

perc_alg = lm.Perceptron()
predictors = ["sepal length", "sepal width", "petal length", "petal width"]
perc_alg.fit(iris_train[predictors],iris_train["species"])

iris_test_preds = perc_alg.predict(iris_test[predictors])
print iris_test_preds
print iris_test["species"]
print metrics.accuracy_score(iris_test["species"],iris_test_preds)



#bin_species = pp.label.binarize(iris_data)
bin_species = pp.label_binarize(iris_data["species"],classes=["Iris-virginica","Iris-setosa","Iris-versicolor"])
print bin_species
bin_species = pandas.DataFrame(bin_species)
print bin_species
iris_data[["Iris-virginica","Iris-setosa","Iris-versicolor"]] = bin_species
print iris_data
```
